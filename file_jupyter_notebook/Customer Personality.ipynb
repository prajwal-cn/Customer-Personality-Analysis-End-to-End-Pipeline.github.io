### Age Group
- age <= 19 : teens
- 19> age >=39 : adults
- 40 > age >= 59 : Middle age adults
- age >= 60 : seniors

```python
df.loc[df['Age'] <=19, "AgeGroup"] = "Teen"
df.loc[(df['Age'] >=20) & (df['Age'] <= 39), "AgeGroup"] = "Adults"
df.loc[(df['Age'] >=40) & (df['Age'] <= 59), "AgeGroup"] = "Middle Age Adults"
df.loc[(df['Age'] >= 60), "AgeGroup"] = "Seniors"
```

### Cluster Modeling

```python
df.columns
```

```python
final_df = df.drop(
    [
      'ID', 'Year_Birth','Education','Marital_Status', 'Kidhome', 
      'Teenhome', 'Dt_Customer', 'MntWines', 'MntFruits', 'MntMeatProducts',
        'MntFishProducts', 'MntSweetProducts','MntGoldProds','NumDealsPurchases',
        'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases','NumWebVisitsMonth',
        'AcceptedCmp3','AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
        'AcceptedCmp2', 'Complain', 'Z_CostContact', 'Z_Revenue', 'Response','AgeGroup'
    ],
    
    axis = 1
)
```

```python
final_df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
model = KMeans()
model.fit(final_df)
```

```python
model.labels_
```

```python
clusterRange = range(2,21)
inertiaRange = []
silhouterange = []

for m in clusterRange:
    model_m = KMeans(n_clusters=m)
    model_m.fit(final_df)
    inertiaRange.append(model_m.inertia_)
    silhouterange.append(metrics.silhouette_score(final_df, model_m.labels_))
```

```python
plt.plot(clusterRange, inertiaRange)
```

```python
print(list(clusterRange))
print("\n")
print(inertiaRange)
print("\n")
print(silhouterange)
```

#### Note: Number of cluster will be 4

```python
model = KMeans(n_clusters=4)
model.fit(final_df)
```
